## **Introduction to OpenCV**

```python
# Importing the required libraries
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# Reading the image
img = cv.imread('coca-cola-logo.png')
assert img is not None, "file could not be read, check with os.path.exists()"
# Displaying the image
cv.imshow("Display window", img)
cv.waitKey(5000)
# Destroying the window after the image is displayed
cv.destroyAllWindows()
```

```python
# Displaying the image using matplotlib
plt.imshow(img)
```

```python
# Displaying the image using matplotlib with rgb color
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```

```python
# Displaying the image using matplotlib with grid lines
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.grid(True)
plt.show()
```

```python
# Printing the shape of the image
print(img.shape)
```

```python
# Printing the data type of the image
print( img.dtype )
```

```python
# Accessing the pixel value at specific coordinates
px = img[100,100]
print( px )
```

```python
# Accessing only red pixel value
red = img[100,100,2]
print( red )
```

```python
# Modifying the pixel value
img[100,100] = [255,255,255]
print( img[100,100] )
```

```python
# accessing RED value
img.item(10,10,2)
```

```python
# modifying RED value
img.itemset((10,10,2),100)
img.item(10,10,2)
```

```python
# Printing the size of the image
print(img.size)
```

## **Image ROI**

```python
coca = img[300:400, 100:200]
img[100:200, 300:400] = coca
```

```python
cv.imshow("Display window", img)
cv.waitKey(3000)
cv.destroyAllWindows()
```

## **Splitting and Merging Image Channels**

```python
# Splitting Image Channels
b, g, r = cv.split(img)

cv.imshow("original", img)
cv.imshow("Blue Channel", b)
cv.imshow("Green Channel", g)
cv.imshow("Red Channel", r)
cv.waitKey(3000)
cv.destroyAllWindows()
```

```python
# Merging Image Channels
merged_image = cv.merge((b, g, r))

cv.imshow("Merged Image", merged_image)
cv.waitKey(3000)
cv.destroyAllWindows()
```

## **Padding**

```python
img = cv.imread('New_Zealand_Lake.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"
```

```python
# Making Borders for Images (Padding)
constant= cv.copyMakeBorder(img,10,10,10,10,cv.BORDER_CONSTANT,value=[255,0,0]) # Blue
replicate= cv.copyMakeBorder(img,10,10,10,10,cv.BORDER_REPLICATE) # Last element is replicated
reflect= cv.copyMakeBorder(img,10,10,10,10,cv.BORDER_REFLECT) # Reflects the elements
reflect101= cv.copyMakeBorder(img,10,10,10,10,cv.BORDER_REFLECT_101) # Reflects the elements
wrap= cv.copyMakeBorder(img,10,10,10,10,cv.BORDER_WRAP) # Wraps the elements

# Displaying the images
cv.imshow("Constant", constant)
cv.imshow("Replicate", replicate)
cv.imshow("Reflect", reflect)
cv.imshow("Reflect101", reflect101)
cv.imshow("Wrap", wrap)
cv.waitKey(3000)
cv.destroyAllWindows()
```

```python
# Displaying all the images using matplotlib
plt.figure(figsize=(20,10))
plt.subplot(231),plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB)),plt.title('ORIGINAL')
plt.subplot(232),plt.imshow(cv.cvtColor(constant, cv.COLOR_BGR2RGB)),plt.title('CONSTANT')
plt.subplot(233),plt.imshow(cv.cvtColor(replicate, cv.COLOR_BGR2RGB)),plt.title('REPLICATE')
plt.subplot(234),plt.imshow(cv.cvtColor(reflect, cv.COLOR_BGR2RGB)),plt.title('REFLECT')
plt.subplot(235),plt.imshow(cv.cvtColor(reflect101, cv.COLOR_BGR2RGB)),plt.title('REFLECT_101')
plt.subplot(236),plt.imshow(cv.cvtColor(wrap, cv.COLOR_BGR2RGB)),plt.title('WRAP')

```
