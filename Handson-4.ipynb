### **Geometric Transformations**

```python
# Importing necessary liabraries
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('building-windows.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"

plt.imshow(img);
```

### **Scaling**

```python
res = cv.resize(img,None,fx=2, fy=2, interpolation = cv.INTER_CUBIC)
#OR
height, width = img.shape[:2]
res = cv.resize(img,(2*width, 2*height), interpolation = cv.INTER_CUBIC)
```

### **Translation**

```python
img1 = cv.imread('building-windows.jpg',cv.IMREAD_GRAYSCALE)
assert img1 is not None, "file could not be read, check with os.path.exists()"
```

```python
rows,cols = img1.shape
M = np.float32([[1,0,100],[0,1,50]])
dst = cv.warpAffine(img,M,(cols,rows))
cv.imshow('img1',dst)
cv.waitKey(1000)
cv.destroyAllWindows()
```

### **Rotation**

```python
#  Rotate the image by 90 degree with respect to center without any scaling.
img =cv.imread('building-windows.jpg')
assert img is not None, "file could not be read, check with os.path.exists()"
rows,cols,ch = img.shape

M = cv.getRotationMatrix2D(((cols-1)/2.0,(rows-1)/2.0),90,1)
dst = cv.warpAffine(img,M,(cols,rows))
cv.imshow('img',dst)
cv.waitKey(1000)
cv.destroyAllWindows()
```

```python
# Affine Transformations
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])
M = cv.getAffineTransform(pts1,pts2)
dst = cv.warpAffine(img,M,(cols,rows))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
# Perspective Transformation

img = cv.imread("pers_tra.jpg")
assert img is not None, "file could not be read, check with os.path.exists()"
rows,cols,ch = img.shape
pts1 = np.float32([[30,240],[640,30],[230,1150],[900,950]])
pts2 = np.float32([[0,0],[1000,0],[0,1200],[1000,1200]])
M = cv.getPerspectiveTransform(pts1,pts2)
dst = cv.warpPerspective(img,M,(1000,1200))
plt.figure(figsize=(10,5))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

### **Image Filtering**


```python
img = cv.imread("Piano_Sheet_Music.png")
assert img is not None, "Image is not found, check with os.path.exists()"
```

```python
kernel = np.ones((5,5),np.float32)/25
dst = cv.filter2D(img,-1,kernel)
plt.figure(figsize=(15,10))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

### **Image Smoothing**

```python
# averaging
blur = cv.blur(img,(5,5))
plt.figure(figsize=(15,10))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Gaussian Blurring
blur = cv.GaussianBlur(img,(5,5),0)
plt.figure(figsize=(15,10))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Gaussian Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Median Blurring
median = cv.medianBlur(img,5)
plt.figure(figsize=(15,10))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(median),plt.title('Median Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Bilateral Filtering
blur = cv.bilateralFilter(img,9,75,75)
plt.figure(figsize=(15,10))
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('Bilateral Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```
